### Read in data from the NASA Exoplanet Archive using their API

```python
import pyvo as vo
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
plt.style.use('presentation')
```

Set up access to the NEA Table Access Protocol

```python
service = vo.dal.TAPService("https://exoplanetarchive.ipac.caltech.edu/TAP")
```

```python
def run_query(query, service=service):
    """ function to execute query """
    
    query_result = service.search(query)  # sync query
    result = query_result.to_table()  # convert to table
    
    return result
```

### Define our query using SQL
We are searching the "ps" table, which corresponds to the <a href="https://exoplanetarchive.ipac.caltech.edu/cgi-bin/TblView/nph-tblView?app=ExoTbls&config=PS">Planetary Systems</a> table. We want to get all of the columns, so will select everything using the asterisk.
We are going to search for table entries where the number of planets (sy_pnum) > 2, where the default flag is 1 (i.e. we are using the preferred parameters in the archive). 

```python
multiplanet_query = """
                    SELECT * 
                    FROM ps
                    WHERE sy_pnum > 5 AND default_flag = 1
                    """
```

```python
# run the query, and convert the output to a Pandas dataframe
result = run_query(multiplanet_query).to_pandas()
```

Let's see how many systems we have

```python
num_systems = result.hostname.unique().shape[0]
# Split into number of rows and columns
print (f"The number of confirmed multi-planet systems is {num_systems}")
```
